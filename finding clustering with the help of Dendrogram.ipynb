```python
#The dendrogram is statistical structure that resembles a tree and is used to depict the 
#relative length relationship among various observation
```

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
```

```python
x,y = make_blobs(n_samples=300,centers=6,cluster_std=0.6)
```

```python
x1=pd.DataFrame(x)
```

```python
x1
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
import scipy.cluster.hierarchy as sch
```

```python
sch.dendrogram(sch.linkage(x))
```

1.first of all extend all horizontal lines of the dendrogram in both the directions
  2.after that find out the largest uncut vertical line
  3.pass a hypothetical infinite plane through that longest uncut vertical line
  4.The number of points at which the hypothetical plane is intersected by the vertical line will give you the optimal number of cluster

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hca=AgglomerativeClustering(n_clusters=7)
```

```python
hca
```

```python
y_pridict=hca.fit_predict(x)
```

```python
y_pridict
```

```python
plt.scatter(x[y_pridict==0,0],x[y_pridict==0,1])
plt.scatter(x[y_pridict==1,0],x[y_pridict==1,1])
plt.scatter(x[y_pridict==2,0],x[y_pridict==2,1])
plt.scatter(x[y_pridict==3,0],x[y_pridict==3,1])
plt.scatter(x[y_pridict==4,0],x[y_pridict==4,1])
plt.scatter(x[y_pridict==5,0],x[y_pridict==5,1])
```

If assume the number of clusters=2 then see the result
